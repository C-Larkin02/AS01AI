```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Define the parameters of the genetic algorithm
mutation_rate = 0.3
crossover_rate = 0.7
population_size = 100
max_generations = 1000

# Read in the data from the provided text file
with open("Binpacking-2.txt", "r") as file:
    text = file.read()

# Split the data on return characters
lines = text.split('\n')

instances = []
instance = {}

# Loop through the file lines
for line in lines:
    # Check for BPP, this means new problem
    if line.startswith("'BPP"):
        # If there's a current problem instance, add it to the list of instances
        if instance:
            instances.append(instance)
        # Start a new problem instance
        instance = {'name': line.strip(), 'weights': [], 'counts': []}
    elif instance:
        # If there's a current problem instance, add the data to it
        if 'num_weights' not in instance:
            instance['num_weights'] = int(line.strip())
        elif 'capacity' not in instance:
            instance['capacity'] = int(line.strip())
        else:
            # This is a line with an item weight and count
            values = line.split()
            if values:
                if len(values) >= 2:
                    weight, count = map(int, values)
                    instance['weights'].append(weight)
                    instance['counts'].append(count)
    
# Add the last problem instance to the list of instances
if instance:
    instances.append(instance)

# Process each problem instance
for instance in instances:
    problem_name = instance['name']
    bin_capacity = instance['capacity']
    item_weights = instance['weights']
    print(item_weights)
    item_counts = instance['counts']
    print(item_counts)
    num_weights = len(item_weights)
    total_items = []
    # Put items into the list for each of their counts, to reate total items list
    for weight, count in zip(item_weights, item_counts):
        total_items.extend([weight] * count)

    print(total_items)

    # Initialising the population using NumPy
    num_bins = len(total_items)
    num_items = len(total_items)
    population = np.random.randint(0, num_bins, (population_size, num_items))
    
   
    def fitness(solution):
        # dictionary to store total weight of each bin
        bin_weights = {}

        # Loop throuh the solution, calculating the total weight of each bin
        bins_used = 0
        for item, bin in enumerate(solution):
            if bin not in bin_weights:
                bin_weights[bin] = 0
                bins_used += 1
            bin_weights[bin] += total_items[item]

        # This line creates a penlalty value for punishing solutions that exceed bin capacity
        penalty = sum(max(0, weight - 1000) for weight in bin_weights.values())

        # add the number of bins used and the penalty to get the final fitness value
        fitness = -bins_used - penalty

        return fitness

    # Mutation function
    def mutate(individual, mutation_rate):
        for i in range(len(individual)):
            if random.random() < mutation_rate:
                individual[i] = random.randint(0, num_bins - 1)

    # Function to calculate the number of bins used by a solution
    def bins_used(solution):
        return len(set(solution))
    
    # Function to check if a solution is feasible
    def is_feasible(solution):

        bin_weights = {}

        for item, bin in enumerate(solution):
            if bin not in bin_weights:
                bin_weights[bin] = 0
            bin_weights[bin] += total_items[item]

        for bin, weight in bin_weights.items():
            if weight > 1000:
                return False

            return True

    # Crossover function
    def crossover(parent1, parent2, crossover_rate):
        if random.random() < crossover_rate:
            crossover_point = random.randint(0, len(parent1) - 1)
            parent1[crossover_point:], parent2[crossover_point:] = parent2[crossover_point:], parent1[crossover_point:]

    # Array for max fitness values
    max_fitness_values = []

    # Aray for average fitness values
    avg_fitness_values = []

    
    # Choose the number of elites
    num_elites = 5
    
    # Next, the genetic algorithm starts

    generation = 0
    while generation < max_generations:
        # Loop through the population and calculate the fitness of each individual
        scores = np.array([fitness(individual) for individual in population])

        # append mean fitness value
        avg_fitness_values.append(np.mean(scores))

        # append max fitness value
        max_fitness_values.append(max(scores))

        # Selecting individuals to reproduce based on their fitness
        scores = np.array([fitness(individual) for individual in population])

        # Identifying the elites (best individuals)
        elites_indices = np.argsort(scores)[-num_elites:]
        elites = population[elites_indices]

        probabilities = scores / scores.sum()
        population = population[np.random.choice(population_size, size=population_size, p=probabilities)]

        # Applying crossover and mutation to create the next generation
        for i in range(num_elites, population_size - 1, 2):
            parent1, parent2 = population[i], population[i + 1]
            crossover(parent1, parent2, crossover_rate)
            mutate(parent1, mutation_rate)
            mutate(parent2, mutation_rate)

        # Copying the elites to the next generation
        population[:num_elites] = elites    

        generation += 1

    # Finding the best solution after the last generation
    best_solution = population[np.argmax(scores)]

    # Calculating the number of bins used by the best solution
    bins_used = bins_used(best_solution)

    print(f"For {problem_name}, With {len(total_items)} items. Number of bins used was: {bins_used}")

    # Checking if the best solution is feasible
    if is_feasible(best_solution):
        print(f"For {problem_name}, the best solution is feasible.")
    else:
        print(f"For {problem_name}, the best solution is not feasible.")
 
    print(f"For {problem_name}, Best solution: {best_solution}")
    print(f"For {problem_name}, Best fitness: {fitness(best_solution)}")

    # Plotting the evolution of the average fitness value
    plt.plot(avg_fitness_values)
    plt.title(f'Evolution of Average Fitness for {problem_name}')
    plt.xlabel('Generation')
    plt.ylabel('Average Fitness')
    plt.show()
```
